#Model building this section will be used to build the logistic regression model
Y axis will be the winpercent 
X axis will be the remaining columns

```python
X = candy.drop(columns='chocolate')
#Assign X variables
y= candy['chocolate']
#set y to the dropped column
from sklearn.preprocessing import LabelEncoder 
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3)
```

```python
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LogisticRegression
```

```python
onehot = OneHotEncoder()
onehot.fit(X[['competitorname','fruity','caramel','peanutyalmondy','nougat','crispedricewafer','hard','bar','pluribus','sugarpercent','pricepercent','winpercent']])
#Encode the data
```

```python
columntransform = make_column_transformer((OneHotEncoder(categories=onehot.categories_),['competitorname','fruity','caramel','peanutyalmondy','nougat','crispedricewafer','hard','bar','pluribus','sugarpercent','pricepercent','winpercent']),remainder='passthrough')
```

```python
logicregress = LogisticRegression()
```

```python
pipe = make_pipeline(columntransform,logicregress)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_predict=pipe.predict(X_test)
#Predicted the 30% percent of the data set, which will be based on the fitted data used in the train set 
```

```python
r2_score(y_test,y_predict)
#Use r2 score to evaluate the accuracy of the model
```

After running training the model we see that the accuracy is 18% how can we imporve this I will be exploring this below

```python
score = []
#Create an array which will store the array scores of the iteration below
for i in range(10):
    #Use 10000 iterations to find the highest accuracy score
    X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=i)
    #Create a train test split of the data above
    regre=LogisticRegression()
    #Create a LogisticRegression variable
    pipe = make_pipeline(columntransform,regre)
    #Create a pipeline that will consist of the column transformer and the logistic regression
    pipe.fit(X_train,y_train)
    #Fit the training data to the model
    y_predict=pipe.predict(X_test)
    #Predict the test data 
    score.append('%.3f'% r2_score(y_test,y_predict))
    #Append the r2 scores into score
    
```

```python
print(np.argmax(score))
#Outout the highest accuracy score location
score[np.argmax(score)] 
#Show the specific r2 score
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=np.argmax(score))
regre = LogisticRegression()
pipe = make_pipeline(columntransform,regre)
pipe.fit(X_train,y_train)
y_predict=pipe.predict(X_test)
r2_score(y_test,y_predict)
#This section will retrain the model and use the greatest accuracy score as its random state which will be used in shuffling before splitting the data
#After fitting the new dataa and predicting the testing we reevaluate the accuracy score and see that the accuracy is at 100%
```
