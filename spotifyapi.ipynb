```python
#call spotify 
import spotipy
import json
#import spotify authentication
from spotipy.oauth2 import SpotifyClientCredentials
import pandas as pd
```

```python
#use my client id and secret
client_id = ''
client_secret = ''
```

```python
client_credentials_manager = SpotifyClientCredentials(client_id=client_id, client_secret=client_secret)
spo = spotipy.Spotify(client_credentials_manager=client_credentials_manager)
playlist_1 = 'https://open.spotify.com/playlist/7063ZiWV0nJgaD42txrVNa?si=2a6470ad71974ef0'
rs= spo.playlist(playlist_1)
resutlt = spo.playlist(playlist_1)
print(resutlt)
```

After calling the api and importing my playlist i would like to convert the JSON file into a DataFrame File

```python
songiD =[]
for item in resutlt['tracks']['items']:
    songiD.append(item['track']['id'])

songmeta = {'id':[],'album':[],'name':[],'artist':[],'popularity':[],'explicit':[]}

for song_ID in songiD:
    #get song metadata
    meta  = spo.track(song_ID)
    #get the song id
    songmeta['id'].append(meta['id'])
    #get the album name
    songmeta['album'].append(meta['album']['name'])
    #get the song name
    songmeta['name'].append(meta['name'])
    #get the artist name
    songmeta['artist'].append(meta['artists'][0]['name'])
    #get the popularity
    songmeta['popularity'].append(meta['popularity'])
    #get the explicit
    songmeta['explicit'].append(meta['explicit'])
    features = spo.audio_features(songmeta['id'])
# change dictionary to dataframe
    features_df=pd.DataFrame.from_dict(features)
#create a dataframe
df = pd.DataFrame(songmeta)
```
